```python
#调用node2vec库函数测试
import networkx as nx
from node2vec import Node2Vec
from matplotlib import pyplot as plt
```

```python
#create a graph
test_graph = nx.fast_gnp_random_graph(n=100,p=0.5) 
nx.draw(test_graph)
```

```python
test_graph.edges() 
```

```python
node2vec_testgraph = Node2Vec(test_graph,dimensions=64,walk_length=30,num_walks=200,workers=4)
```

```python
model = node2vec_testgraph.fit(window=10,min_count=1,batch_words=4)
```
