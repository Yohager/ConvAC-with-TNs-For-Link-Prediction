```python
import numpy as np
import random
import networkx as nx
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#加载nx库中的自带数据
G_karate = nx.karate_club_graph()
```

```python
#绘制G_karate图像
nx.draw(G_karate,cmap=plt.get_cmap('rainbow'),with_labels=True,pos=nx.spring_layout(G_karate))
```

```python
n = G_karate.number_of_nodes()
m = G_karate.number_of_edges()
print("Number of nodes :", str(n))
print("Number of edges :", str(m))
print("Number of connected components :" ,str(nx.number_connected_components(G_karate)))
```

```python
#划分数据集，从原图中删除25%的连边后作为训练数据
edges_subset = random.sample(G_karate.edges(),int(0.25 * G_karate.number_of_edges()))
G_karate_train = G_karate.copy()
G_karate_train.remove_edges_from(edges_subset)
```

```python
plt.figure(figsize=(8,6))
nx.draw(G_karate_train)
```

```python
prediction_jaccard = list(nx.jaccard_coefficient(G_karate_train))
```

```python
prediction_jaccard
```

```python
score,label_1,label_2 = zip(*[(s,(u,v) in edges_subset,(u,v) in G_karate.edges()) for (u,v,s) in prediction_jaccard])
```

```python
data_all = list(zip(score,label_1,label_2))
```

```python
#AUC函数
def auc_function(score,label,times):
    #大致思路是从data_test中选出一条边，然后从score中得到其分数
    #然后选择score list中的另一个边，同时这个边不在训练数据和测试数据中的bian
    #比较两个边的分数，如果从data_test中选出的边的分数高则加1分，分数相同则加0.5分，分数低了加0分
    #比较次数参数为times
    compare_times = times
    data_all = list(zip(score,label))
    data_true = []
    data_false = []
    for i in data_all:
        if i[1] == True:
            data_true.append(i)
        else:
            data_false.append(i)
    val_1 = 0
    while (compare_times > 0):
        data_1 = random.choice(data_true)
        data_2 = random.choice(data_false)
        if (data_1[0] > data_2[0]):
            val_1 += 1
        elif (data_1[0] == data_2[0]):
            val_1 += 0.5
        else:
            pass
        compare_times -= 1
    return val_1/times
```

```python
auc = auc_function(score,label_1,100)
```

```python
auc
```

```python
def precision_function(score,label,L):
    data_merge = list(zip(score,label))
    data_merge.sort(reverse=True)
    count = 0
    for i in range(L):
        if (data_merge[i][1] == True):
            count += 1
    return count/L
```

```python
precision_result = precision_function(score,label_1,200)
```
